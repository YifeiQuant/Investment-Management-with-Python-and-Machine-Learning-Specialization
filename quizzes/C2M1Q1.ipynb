```python
import numpy as np
import pandas as pd
import risk_kit as rk
import statsmodels.api as sm
%load_ext autoreload
%autoreload 2
```

```python
ind = rk.get_ind_returns(weighting='vw', n_inds=49)
fff = rk.get_fff_returns()
```

```python
ind.head()
```

```python
fff.head()
```

```python
ind_excess = ind.loc['1991':'2018'] - fff.loc['1991':'2018', ['RF']].values
mkt_excess = fff.loc['1991':'2018', ['Mkt-RF']]
exp_var = mkt_excess.copy()
exp_var['Constant'] = 1
lm = sm.OLS(ind_excess['Beer'], exp_var).fit()
lm.summary()
```

```python
lm.params['Mkt-RF'].round(2)
```

```python
lm_steel = sm.OLS(ind_excess['Steel'], exp_var).fit()
lm_steel.params['Mkt-RF'].round(2)
```

```python
ind_excess = ind.loc['2013':'2018'] - fff.loc['2013':'2018', ['RF']].values
mkt_excess = fff.loc['2013':'2018', ['Mkt-RF']]
exp_var = mkt_excess.copy()
exp_var['Constant'] = 1
lm_beer_13_18 = sm.OLS(ind_excess['Beer'], exp_var).fit()
lm_beer_13_18.params['Mkt-RF'].round(2)
```

```python
lm_steel_13_18 = sm.OLS(ind_excess['Steel'], exp_var).fit()
lm_steel_13_18.params['Mkt-RF'].round(2)
```

```python
ind_excess = ind.loc['1991':'1993'] - fff.loc['1991':'1993', ['RF']].values
mkt_excess = fff.loc['1991':'1993', ['Mkt-RF']]
exp_var = mkt_excess.copy()
exp_var['Constant'] = 1
```

```python
ind_betas = {industries: sm.OLS(ind_excess[industries], exp_var).fit().params['Mkt-RF'] for industries in ind_excess.columns}
ind_betas
```

```python
max(ind_betas, key=ind_betas.get)
```

```python
min(ind_betas, key=ind_betas.get)
```

```python
ind_excess = ind.loc['1991':'2018'] - fff.loc['1991':'2018', ['RF']].values
exp_var = fff.loc['1991':'2018', ['Mkt-RF', 'SMB', 'HML']]
size_coef = {industries : rk.regress(ind_excess[industries], exp_var).params['SMB'] for industries in ind_excess.columns}
```

```python
min(size_coef, key=size_coef.get)
```
