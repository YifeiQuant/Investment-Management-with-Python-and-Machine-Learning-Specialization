```python
import pandas as pd
import numpy as np
import risk_kit as rk
```

```python
me_m = pd.read_csv('data/Portfolios_Formed_on_ME_monthly_EW.csv',
                    header = 0, index_col = 0, 
                    parse_dates = True, na_values = -99.99
                    )
rets = me_m[['Lo 20', 'Hi 20']] / 100
rets.columns = ['SmallCap', 'LargeCap']
rets.index = pd.to_datetime(rets.index, format = '%Y%m').to_period('M')
rets.head()
```

## Q1 & Q3

```python
annualized_rets = (((rets + 1).prod()**(12 / rets.shape[0]) - 1) * 100).round(2)
annualized_rets
```

## Q2 & Q4

```python
annulized_vol = (rets.std(ddof = 0) * np.sqrt(12) * 100).round(1) 
annulized_vol
```

## Q5 & Q7

```python
annualized_rets = (((rets.loc['1999':'2015'] + 1).prod()**(12 / rets.loc['1999':'2015'].shape[0]) - 1) * 100).round(2)
annualized_rets
```

```python
annulized_vol = (rets.loc['1999':'2015'].std(ddof = 0) * np.sqrt(12) * 100).round(1) 
annulized_vol
```

## Q9 & Q10

```python
maxDD = -rk.drawdown(rets.loc['1999':'2015']['SmallCap'])['Drawdowns'].min() * 100
maxDD = round(maxDD, 2)
maxDD
```

```python
maxDD_date = rk.drawdown(rets.loc['1999':'2015']['SmallCap'])['Drawdowns'].idxmin()
maxDD_date
```

## Q11 & Q12

```python
maxDD = -rk.drawdown(rets.loc['1999':'2015']['LargeCap'])['Drawdowns'].min() * 100
maxDD = round(maxDD, 2)
maxDD
```

```python
maxDD_date = rk.drawdown(rets.loc['1999':'2015']['LargeCap'])['Drawdowns'].idxmin()
maxDD_date
```
