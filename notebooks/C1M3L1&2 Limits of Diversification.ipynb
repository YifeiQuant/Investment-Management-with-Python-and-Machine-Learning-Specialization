```python
%load_ext autoreload
%autoreload 2

import risk_kit as rk
import numpy as np
import pandas as pd
```

```python
ind_rets = rk.get_ind_returns()
ind_nfirms = rk.get_ind_nfirms()
ind_size = rk.get_ind_size()
```

```python
ind_rets.shape
```

```python
ind_nfirms.shape
```

```python
ind_size.shape
```

```python
ind_nfirms.head()
```

```python
ind_mktcap = ind_nfirms * ind_size
ind_mktcap.shape
```

```python
total_mktcap = ind_mktcap.sum(axis = 'columns')
total_mktcap.plot(figsize = (12, 6))
```

```python
ind_capweight = ind_mktcap.divide(total_mktcap, axis = 'rows')
```

```python
ind_capweight.head()
```

```python
ind_capweight.loc['1926'].sum(axis = 'columns')
```

```python
ind_capweight[['Fin', 'Steel']].plot(figsize = (12, 6))
```

```python
total_market_return = (ind_capweight * ind_rets).sum(axis = 'columns')
total_market_return.head() 
```

```python
total_market_return.plot()
```

```python
total_market_index = rk.drawdown(total_market_return).Wealth
total_market_index.plot(figsize = (10, 5), title = 'Total Market Cap-Weighted Index 1926-2018')
```

```python
total_market_index.loc['1980':].plot(figsize = (12, 6))
total_market_index.loc['1980':].rolling(window = 36).mean().plot()
```

```python
tmi_tr36rets = total_market_return.rolling(window = 36).aggregate(rk.annualize_rets, periods_per_year = 12)
tmi_tr36rets.plot(figsize = (12, 6), label = 'Tr36 mo Returns', legend = True)
total_market_return.plot(label = 'Returns', legend = True)
```

### Rolling Correlation - along with MultiIndexes and .groupby

```python
ts_corr = ind_rets.rolling(window = 36).corr()
ts_corr.tail()
```

```python
ts_corr.index.names = ['Date', 'Industry']
```

```python
ts_corr.tail()
```

```python
ind_tr36corr = ts_corr.groupby(level = 'Date').apply(lambda cormat: cormat.values.mean())
ind_tr36corr
```

```python
ind_tr36corr.plot()
```

```python
tmi_tr36rets.plot(label = 'Tr36 Mo Rets', figsize = (12, 6), legend = True)
ind_tr36corr.plot(label = 'Tr36 Mo Corr', legend = True, secondary_y = True)
```

```python
tmi_tr36rets.loc['2008':].plot(label = 'Tr36 Mo Rets', figsize = (12, 6), legend = True)
ind_tr36corr.loc['2008':].plot(label = 'Tr36 Mo Corr', legend = True, secondary_y = True)
```
